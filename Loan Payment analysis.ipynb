```python
df= df.drop(['Loan_ID','effective_date','due_date','paid_off_time'], axis='columns')
df
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
df['loan_status']=labelencoder.fit_transform(df['loan_status'])
df['education']=labelencoder.fit_transform(df['education'])
df['Gender']= pd.get_dummies(df['Gender'])
df
```

```python
data=df.drop(labels='loan_status', axis=1)
target=df.loan_status
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data, target, test_size=0.2)
```

```python
%%time
from sklearn.linear_model import LogisticRegression
lr= LogisticRegression()
lr.fit(x_train, y_train)
lr.score(x_test, y_test)
```

```python
lr_predict=lr.predict(x_test)
from sklearn.metrics import confusion_matrix
cm_lr= confusion_matrix(y_test, lr_predict)
import seaborn as sns
plt.figure(figsize=(12,7))
sns.heatmap(cm_lr, annot= True, Cmap= 'Accent')
plt.xlabel('predicted')
plt.ylabel('Truth')
```

```python
%%time
from sklearn.ensemble import RandomForestClassifier
rf= RandomForestClassifier()
rf.fit(x_train, y_train)
rf.score(x_test, y_test)
```

```python
rf_predict=rf.predict(x_test)
from sklearn.metrics import confusion_matrix
cm_rf= confusion_matrix(y_test, rf_predict)
import seaborn as sns
plt.figure(figsize=(12,7))
sns.heatmap(cm_rf, annot= True, Cmap= 'Accent')
plt.xlabel('predicted')
plt.ylabel('Truth')
```

```python
%%time
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier()
tree.fit(x_train,y_train)
tree.score(x_test, y_test)
```

```python
tree_predict=tree.predict(x_test)
from sklearn.metrics import confusion_matrix
cm_tree= confusion_matrix(y_test, tree_predict)
import seaborn as sns
plt.figure(figsize=(12,7))
sns.heatmap(cm_tree, annot= True)
plt.xlabel('predicted')
plt.ylabel('Truth')
```

```python
model_X= ['LogisticRegression', 'Decision Tree', 'Random Forest']
model_Y= [lr.score(x_test,y_test),tree.score(x_test,y_test),rf.score(x_test,y_test)]
plt.figure(figsize=(12,7))
plt.plot(model_X, model_Y, color='green', marker='*', linestyle='dashed')
plt.title('model performance')
plt.xlabel('model')
plt.ylabel('model accuracy')
```

```python
from sklearn.svm import SVC
model_params={
    'svm':{
        'model':SVC(gamma='auto'),
        'params':{
        'C':[1,10,20],
        'kernel':['rbf','linear']
         }
    },
    'random_forest':{
        'model':RandomForestClassifier(),
        'params':{
            'n_estimators':[1,5,10]
        }
    },
    'logistic_regression':{
        'model':LogisticRegression(solver='liblinear',multi_class='auto'),
        'params':{
            'C':[1,5,10]
        }
    },
    'decision_tree':{
        'model':RandomForestClassifier(),
        'params':{
            'criterion':['gini','entropy']
        }
        
    }
        
}
```

```python

from sklearn.model_selection import GridSearchCV

scores=[]
for model_name,mp in model_params.items():
    clf= GridSearchCV(mp['model'],mp['params'],cv=3,return_train_score=False)
    clf.fit(data,target)
    scores.append({
        'model':model_name,
        'best_score':clf.best_score_,
        'best_params':clf.best_params_
    })
```

```python
df=pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```
